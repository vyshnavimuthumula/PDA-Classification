## Training and Test data

```python
# Selecting input data
X = smoking_df[['systolic','serum creatinine','fasting blood sugar','triglyceride','Urine protein','hemoglobin','AST','LDL','relaxation','Gtp','dental caries']].values
```

```python
# Defining output variable
y = smoking_df['smoking'].values
```

```python
# Polynomial feature with degree 3
poly_conv=PolynomialFeatures(degree=3)
```

```python
# Transform input data
poly_feat=poly_conv.fit_transform(X)
```

```python
# Calculating zscore for normalising the data
X=stats.zscore(X)
X[0:5]
```

**Training and Testing data set**

```python
# evaluation of a model using 5 features which generates clusters which 11 columns and 1000 rows randomly

X, y = make_classification(n_samples=1000, n_features=11, n_informative=5, n_redundant=5, random_state=42)
# split into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)

print ('Train set:', X_train.shape,  y_train.shape)
print ('Test set:', X_test.shape,  y_test.shape)
```

## Logistic Regression

```python
# Fit Logistic regression model 
LR = LogisticRegression().fit(X_train,y_train)
```

```python
# Predecting the test data in binary format.
smoke_lr_pred = LR.predict(X_test)
smoke_lr_pred
```

```python
# Predecting the test data.
smoke_yhat_prob = LR.predict_proba(X_test)
smoke_yhat_prob
```

```python
# Finding accuracy score for logistic regression algorithm using accuracy_score function
smoke_lr_acc = metrics.accuracy_score(y_test, smoke_lr_pred)
print('Logistic Regression accuracy score: %.2f' % (smoke_lr_acc*100))
```

## Random Forest Classifier

```python
# Creating Random Forest Classifier model
rfc = RandomForestClassifier(random_state=42)
```

```python
# Fitting the train data.
rfc.fit(X_train, y_train)
```

```python
#Â PredictÂ theÂ input test data
rfc_pred = rfc.predict(X_test)
```

```python
# Accuracy score by using Random Forest Classifier
smoke_rfc_acc = accuracy_score(y_test, rfc_pred)
print('Random forest Model accuracy score :',smoke_rfc_acc * 100)
```

## K nearest neighbor (K-NN)

**Training and finding the nearst neighbors**

```python
# setting k value
arr_K = 20
smoke_mean_acc = np.zeros((arr_K-1))
for n in range(1,arr_K):
    
    #Train the Model and Predict it
    smoke_neigh = KNeighborsClassifier(n_neighbors = n).fit(X_train,y_train)
    # predecting the input test data
    smoke_pred=smoke_neigh.predict(X_test)
    # calculating accuracy score
    smoke_mean_acc[n-1] = metrics.accuracy_score(y_test, smoke_pred)
  
# printing mean absolute error for KNN
smoke_mean_acc
```

```python
# Ploting the graph accuracy versus k-neighbours
plt.title('Accuracy of the set by using KNN')
plt.xlabel('Number of K-Neighbours')
plt.ylabel('Accuracy score')
plt.tight_layout()
sns.set_theme(style="darkgrid")
plt.plot(range(1,arr_K),smoke_mean_acc)
plt.show()
```

From the prediction we can conclude that at K=9 the accuracy score is high.

```python
# initializing the k value
k = 9
#Fit the trained data set
smoke_neigh = KNeighborsClassifier(n_neighbors = k).fit(X_train,y_train)
```

**Predicting the output**

```python
# Predecting the output
smoke_predict = smoke_neigh.predict(X_test)
smoke_predict[0:5]
```

```python
# Finding the accuracy score using inbuild function
smoke_knn_acc = metrics.accuracy_score(y_test, smoke_predict)*100
print("Smoke Test data set Accuracy with KNN: ", smoke_knn_acc)
```

The accuracy for smoke data is 92.4% when K=9.

```python
# Plotting the graph between test and predicted data
plt.plot( y_test, smoke_predict)
plt.title('Graph between test and predection data')
plt.ylabel('Prediction')
plt.xlabel('Test data')
```

```python
 # Plotting the graph between test data and prediction data
plt.title('Accuracy of the set by using KNN')
plt.xlabel('Number of K-Neighbours')
plt.ylabel('Accuracy score')
plt.tight_layout()
sns.set_theme(style="darkgrid")
plt.plot(y_test,smoke_predict)
plt.show()
```

# Feature Importance for the inputs with Machine learning models

```python
# Fitting the model
svc =  svm.SVC(kernel='rbf', C=10)
svc.fit(X_train, y_train)
# Calculating feature imporatnce by using permutation for model with X_test and y_test data
smoking_perm_imprt = permutation_importance(svc, X_test, y_test)

# Assigning list of input features to a variable so that we can directly use that variable for calculation
smoking_features_labl = ['systolic','serum creatinine','fasting blood sugar','triglyceride','Urine protein','hemoglobin','AST','LDL','relaxation','Gtp','dental caries']
# Converting list of items into numpy array
features_arr = np.array(smoking_features_labl)

# Calculating importance mean and sorting it then stored into a variable
feature_imp = smoking_perm_imprt.importances_mean.argsort()
# Assignining importance mean for the sorted value into a variable called 'a'
a = smoking_perm_imprt.importances_mean[feature_imp]
# Plotting a horizontal graph for feature importance versus mean of the feature importance so that highest feature importance value can be found
plt.barh(features_arr[feature_imp], a)
# Adding titles, x and y labels
plt.title("Feature Importance Graph for the smoking data Inputs", fontsize=20)
plt.xlabel("Permutation Importance")
plt.ylabel("Feature inputs")
plt.show()
```
